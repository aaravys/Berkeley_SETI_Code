```python
import pylab as plt
```

```python
from blimpy import Waterfall 
```

```python
#Let's take a look at the basic information, using the handy info function. 
```

```python
obs = Waterfall('voyager_f1032192_t300_v2.fil')
obs.info()
```

```python
#Now, we can look at the header and some other data.
```

```python
print(obs.header)
print(obs.data.shape)
```

```python
#now we plot out the data. This is  the visualized data of Voyager 1. 
```

```python
obs.plot_spectrum()
```

```python
#now, we can zoom in on the voyager graph to see some narrowband signals. 
```

```python
obs.plot_spectrum(f_start=8420.18, f_stop=8420.25)
```

```python
#In the next step, we plot the graph again, but this time we zoom in a little bit more to examine the graph's features more closely. 
```

```python
plt.figure(figsize=(8, 6))
plt.subplot(3,1,1)
obs.plot_spectrum(f_start=8420.193, f_stop=8420.195) # left sideband
plt.subplot(3,1,2)
obs.plot_spectrum(f_start=8420.2163, f_stop=8420.2166) # carrier
plt.subplot(3,1,3)
obs.plot_spectrum(f_start=8420.238, f_stop=8420.24) # right sideband
plt.tight_layout()
```

```python
#Finally, we construct a waterfall plot. 
```

```python
obs.plot_waterfall(f_start=8420, f_stop=8420.7)
```
